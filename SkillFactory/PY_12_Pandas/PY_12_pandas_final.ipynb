```python
import pandas as pd

ratings_df = pd.read_csv('data/ratings_movies.csv')
ratings_df.head()
```

```python
#библиотека для регулярных выражений
import re 
def get_year_release(arg):
    #находим все слова по шаблону "(DDDD)"
    candidates = re.findall(r'\(\d{4}\)', arg) 
    # проверяем число вхождений
    if len(candidates) > 0:
        #если число вхождений больше 0,
	#очищаем строку от знаков "(" и ")"
        year = candidates[0].replace('(', '')
        year = year.replace(')', '')
        return int(year)
    else:
        #если год не указан, возвращаем None
        return None
```

```python
# 8.1
# Создайте в таблице новый признак year_release, который соответствует году выпуска фильма.
# У скольких фильмов не указан год их выпуска?
ratings_df['year_release'] = ratings_df['title'].apply(get_year_release)
#ratings_df['year_release'] = ratings_df['year_release'].astype('int64')
#print(int(ratings_df['year_release'][0]))
```

```python
print(ratings_df['year_release'].count())
ratings_df.info()
```

```python
# 8.2
# Какой фильм, выпущенный в 1999 году, получил наименьшую среднюю оценку зрителей?
#В качестве ответа запишите название этого фильма без указания года его выпуска.
mask = ratings_df['year_release'] == 1999.0
group_rating = ratings_df[mask].groupby(['title'])['rating'].mean().idxmin()
#group_rating = ratings_df[mask].groupby(['title'])['rating'].mean().sort_values()
display(group_rating)

# Bloodsport: The Dark Kumite
```

```python
# 8.3
# Какое сочетание жанров фильмов (genres), выпущенных в 2010 году, 
# получило наименьшую среднюю оценку (rating)?
# Запишите сочетание так же, как оно указано в таблице (через разделитель |, без пробелов).

mask = ratings_df['year_release'] == 2010.0
group_rating = ratings_df[mask].groupby(['genres'])['rating'].mean().idxmin()
display(group_rating)

# Action|Sci-Fi
```

```python
# 8.4
# Какой пользователь (userId) посмотрел наибольшее количество 
# различных (уникальных) жанров (genres) фильмов? 
# В качестве ответа запишите идентификатор этого пользователя.

group_rating = ratings_df.groupby(['userId'])['genres'].nunique().idxmax()
display(group_rating)

# 599
```

```python
# 8.5
# Найдите пользователя, который выставил наименьшее количество оценок, 
# но его средняя оценка фильмам наибольшая.
# В качестве ответа укажите идентификатор этого пользователя.

group_rating = ratings_df.groupby(by = 'userId')['rating'].agg(
    ['count', 'mean']
).sort_values(by = ['count', 'mean'], ascending= [True, False])
display(group_rating)

# 53
```

```python
# 8.6
# Найдите сочетание жанров (genres) за 2018 году, которое имеет наибольший средний рейтинг 
# (среднее по столбцу rating), и при этом число выставленных ему оценок 
# (количество значений в столбце rating) больше 10.
# Запишите сочетание так же, как оно указано в таблице (через разделитель |, без пробелов).

mask = ratings_df['year_release'] == 2018.0
group_rating = ratings_df[mask].groupby('genres')['rating'].agg(
    ['mean', 'count']
).sort_values(by = ['count', 'mean'], ascending= [False, False])
display(group_rating)

# Action|Adventure|Sci-Fi
```

```python
# 8.7
# Добавьте в таблицу новый признак year_rating — год выставления оценки. 
# Создайте сводную таблицу, которая иллюстрирует зависимость среднего рейтинга фильма 
# от года выставления оценки и жанра. 

ratings_df['date'] = pd.to_datetime(ratings_df['date'])
ratings_df['year_rating'] = ratings_df['date'].dt.year
pivot = ratings_df.pivot_table(
    values = 'rating',
    index = 'genres',
    columns = 'year_rating',
    aggfunc= 'mean'
)

display(pivot.loc['Action|Adventure'].min()) # минимальная оценка жанра Action|Adventure

display(pivot.loc['Action|Adventure|Animation|Children|Comedy|IMAX'].max()) # максимальная оценка жанра Action|Adventure|Animation|Children|Comedy|IMAX
display(pivot.loc['Action|Adventure|Animation|Children|Comedy|IMAX'][2010]) # смотрим, была ли в 2010 году максимальная оценка

display(pivot[2018].sort_values(ascending= False).head(30))
display(pivot.loc['Comedy'])
```

```python
# Дано два файла:

# orders.csv, содержащий данные о заказах;
# products.csv, содержащий данные о товарах.

orders = pd.read_csv('data/orders.csv', sep= ';')
products = pd.read_csv('data/products.csv', sep= ';')

display(orders)
display(products)
```

```python
orders_products = orders.merge(
    products,
    how = 'left',
    left_on = 'ID товара',
    right_on= 'Product_ID',
)
display(orders_products)

# 8.9
# Какой идентификатор (Order ID) имеет заказ, для которого не оказалось информации о товаре?
# 0
```

```python
# 8.9
# На какой товар была произведена отмена?
# В качестве ответа запишите название этого товара (Name).

mask = orders_products['Отменен'] == 'Да'
display(orders_products[mask])

# Носки беговые Camino
```

```python
# 8.10
# Какой покупатель принёс наибольшую суммарную прибыль интернет-магазину за указанный период?
# В ответ запишите идентификатор этого покупателя (ID Покупателя).
# Прибыль состоит только из оплаченных заказов и рассчитывается как количество купленного товара, 
# умноженное на его цену.

mask = orders_products['Оплачен'] == 'Да'
payed_orders = orders_products[mask]
payed_orders['total'] = payed_orders['Количество'] * payed_orders['Price']
payed_orders.groupby('ID Покупателя')['total'].sum().sort_values()
```
