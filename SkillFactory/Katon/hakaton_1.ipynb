```python
df_1 = data_1[['ogrn', 'fullName', 'regionName', 'mainOkved', 'addOkved', 'dateReg', 'dateLiquid', 'incomeTotal', 'incomeDetail']]
```

```python
df_1
```

```python
# фильтруем названия организаций
# 'БЛАГОТВОРИТ', но без “живот”, “зоо”, “собак”, “кошк”, “сельхоз”, “сельскохоз”
df_1['fullName'] = df_1['fullName'].apply(lambda s:str(s).upper())
mask1 = df_1['fullName'].str.contains('БЛАГОТВОРИТ')
mask2 = ~((df_1['fullName'].str.contains('ЖИВОТ')) & (df_1['fullName'].str.contains('ЗОО')) & (df_1['fullName'].str.contains('СОБАК')))
mask3 = ~((df_1['fullName'].str.contains('КОШК')) & (df_1['fullName'].str.contains('СЕЛЬХОЗ')) & (df_1['fullName'].str.contains('СЕЛЬСКОХОЗ')))
 
df_1 = df_1[mask1 & mask2 & mask3]
```

```python
df_1.info()
```

```python
from ast import literal_eval as le
```

```python
# вытаскиваем из mainOkved название деятельности и записываем это в новый столбец mainOkved_name
def okved(x):
   result = le(x)
   return result['name']
df_1['mainOkved_name'] = df_1['mainOkved'].apply(okved)
```

```python
# вытаскиваем из mainOkved код деятельности и записываем это в новый столбец mainOkved_code
def okved(x):
   result = le(x)
   return result['code']
df_1['mainOkved_code'] = df_1['mainOkved'].apply(okved)
```

```python
df_1
```

```python
# вытаскиваем гранты из incomeDetail
def grants_count(x):
   result = le(x)
   return result['grants']['totalCount']
 
def grants_sum(x):
   result = le(x)
   return result['grants']['totalSum']
 
# вытаскиваем субсидии из incomeDetail
def fedSubsidies_count(x):
   result = le(x)
   return result['fedSubsidies']['totalCount']
 
def fedSubsidies_sum(x):
   result = le(x)
   return result['fedSubsidies']['totalSum']
 
# вытаскиваем все контракты из incomeDetail
def contracts_count(x):
   result = le(x)
   count = result['contracts44']['totalCount'] + result['contracts223']['totalCount'] + result['contracts94']['totalCount']
   return count
   
 
def contracts_sum(x):
   result = le(x)
   sum = result['contracts44']['totalSum'] + result['contracts223']['totalSum'] + result['contracts94']['totalSum']
   result = le(x)
   return sum
```

```python
df_1['grants_count'] = df_1['incomeDetail'].apply(grants_count) #количество грантов
df_1['grants_sum'] = df_1['incomeDetail'].apply(grants_sum) #сумма грантов
 
df_1['fedSubsidies_count'] = df_1['incomeDetail'].apply(fedSubsidies_count) #количество субсидий
df_1['fedSubsidies_sum'] = df_1['incomeDetail'].apply(fedSubsidies_sum) #сумма субсидий
 
df_1['contracts_count'] = df_1['incomeDetail'].apply(contracts_count) #количество контрактов
df_1['contracts_sum'] = df_1['incomeDetail'].apply(contracts_sum) #сумма контрактов
```
