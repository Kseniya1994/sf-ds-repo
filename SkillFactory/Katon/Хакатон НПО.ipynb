```python
code_df = data[['ogrn', 'fullName', 'opf', 'oktmo', 'mainOkved']]
code_df.head()
```

```python

code_df['opf_name'] = code_df['opf'].apply(lambda x: le(x)['name'])
code_df['opf_code'] = code_df['opf'].apply(lambda x: le(x)['code'])
code_df['oktmo_name'] = code_df['oktmo'].apply(lambda x: le(x)['name'])
code_df['oktmo_code'] = code_df['oktmo'].apply(lambda x: le(x)['code'])
code_df['mainOkved_name'] = code_df['mainOkved'].apply(lambda x: le(x)['name'])
code_df['mainOkved_code'] = code_df['mainOkved'].apply(lambda x: le(x)['code'])
```

```python
code_df.oktmo.nunique()
```

```python
code_df.ogrn.count()
```

```python
code_df =code_df.drop(['opf'], axis=1)
code_df =code_df.drop(['oktmo'], axis=1)
code_df =code_df.drop(['mainOkved'], axis=1)
```

```python
code_df.head(2)
```

```python
Okved_count =code_df.mainOkved_code.value_counts().reset_index()
Okved_count.rename(columns = {'index': 'Okved', 'mainOkved_code':'count'}, inplace = True)
Okved_count
```
